# Task 2.1: Data Pre-processing

```python
# Importing necessary libraries
import seaborn as sns
from textblob import TextBlob
import sys
from wordcloud import WordCloud
import pandas as pd
import re
import numpy as np
import plotly
import plotly.express as px
import plotly.io as pio
import matplotlib.pyplot as plt
import tweepy,json
import numpy as np
import matplotlib.pyplot as plt
```

## Load the dataset

```python
df = pd.read_csv('Downloads/Dataset-3/Bitcoin_tweets.csv')
df
```

## Filter Tweets with User Locations:

```python
# Assuming 'user_location' is the column containing user locations
df_with_locations = df.dropna(subset=['user_location']).sample(n=500, random_state=42)
```

```python
df_with_locations
```

## Text Preprocessing:

```python
def preprocess_text(text):
    # Remove URLs
    text = re.sub(r'http\S+', '', text)
    # Remove special characters and symbols
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # Remove extra whitespaces
    text = re.sub(r'\s+', ' ', text).strip()
    return text

# Apply preprocessing to the 'text' column
df_with_locations['processed_text'] = df_with_locations['text'].apply(preprocess_text)
```

```python
df_with_locations
```

# Task 2.2: Geocoding

```python
from geopy.geocoders import Nominatim
from tqdm import tqdm

# Initialize the geocoder
geolocator = Nominatim(user_agent="your_geocoding_app_name")

# Create functions to geocode locations
def geocode_location(location):
    try:
        location_info = geolocator.geocode(location)
        if location_info:
            return location_info.latitude, location_info.longitude
        else:
            return None, None
    except Exception as e:
        print(f"Error geocoding {location}: {e}")
        return None, None
```

```python
# Apply geocoding to the 'user_location' column
tqdm.pandas(desc="Geocoding")
df_with_locations[['latitude', 'longitude']] = df_with_locations['user_location'].progress_apply(geocode_location).apply(pd.Series)
```

```python
# Print the DataFrame with added latitude and longitude columns
print(df_with_locations[['user_location', 'latitude', 'longitude']])
```

```python
df_with_locations
```

# Task 2.3 Polarity analysis

## Calculate Polarity:

```python
# Assuming 'processed_text' is the column with preprocessed text
df_with_locations['polarity'] = df_with_locations['processed_text'].apply(lambda x: TextBlob(x).sentiment.polarity)
```

```python
df_with_locations
```

## Average Polarity for Each Location:

```python
average_polarity_by_location = df_with_locations.groupby(['latitude', 'longitude'])['polarity'].mean().reset_index()
```

```python
average_polarity_by_location
```

## Geospatial Visualization:

```python
import folium

# Create a map centered at a specific location
map_center = [average_polarity_by_location['latitude'].mean(), average_polarity_by_location['longitude'].mean()]
folium_map = folium.Map(location=map_center, zoom_start=2)
```

```python
# Add markers for each location with average polarity
for index, row in average_polarity_by_location.iterrows():
    folium.Marker([row['latitude'], row['longitude']],
                  popup=f'Average Polarity: {row["polarity"]:.2f}').add_to(folium_map)
```

```python
# Save the map as an HTML file or display it
folium_map.save('polarity_map.html')
```

- The resulting HTML file (polarity_map.html) will contain the geospatial visualization with markers indicating the average polarity for each location.

- Open the HTML file in a web browser to view the interactive map.
